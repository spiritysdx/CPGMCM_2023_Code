```python
import re
import math
import warnings
import tqdm
import json
import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import optimize


warnings.filterwarnings("ignore")
```

```python
data1 = pd.read_excel('2c治疗方案.xlsx') # 治疗方案
data2 = pd.read_excel('2a数据_已替换时间戳_已矫正.xlsx') # 水肿数据
data3=pd.read_excel('1a数据_已替换时间戳_已矫正.xlsx')# 血肿数据
data3
```

```python
data_w = pd.merge(data1, data2, how='outer', on='ID')
data_w = data_w.drop(range(101, 131))
data_b = pd.merge(data1, data3, how='outer', on='ID')
data_b = data_b.drop(range(101, 131))
```

```python
import json
with open('best_labels_130.json','r') as f:
    groups_label=json.load(f)
print(len(groups_label['best_labels_130']))
```

```python
dataT=data_w
dataT['分组信息']=groups_label['best_labels_130']
dataT.to_excel('含分组信息数据_水肿.xlsx',index=False)
dataH=data_b
dataH['分组信息']=groups_label['best_labels_130']
dataH.to_excel('含分组信息数据_血肿.xlsx',index=False)
```

```python
# 创建一个新列"差"，初始化为0
dataT['差'] = 0
dataH['差'] = 0
U=globals()
resT={}
for i in list(set(dataT['分组信息'])):
    U[f'dataT_g{i}']=dataT[dataT['分组信息']==i]

# 遍历数据集的每一行
    for index, row in U[f'dataT_g{i}'].iterrows():
        row_sum = 0  # 用于累积每行的差值

        prev_value = None  # 用于存储前一个以"ED"开头的列的值

        # 遍历以"ED"开头的列
        for col in row.index:
            if col.startswith('ED'):
                value = row[col]
                if not pd.isna(value):  # 确保元素不是NaN
                    if prev_value is not None:
                        row_sum += (value - prev_value)  # 计算差值并累积
                    prev_value = value

        # 根据row_sum的值设置"差"列的值
        if row_sum > 0:
            U[f'dataT_g{i}'].at[index, '差'] = 1
        elif row_sum < 0:
            U[f'dataT_g{i}'].at[index, '差'] = -1
        resT[f'组{i}']=float(U[f'dataT_g{i}']['差'].sum())

res_resT = {}
for i in resT:
    if resT[i] < 0:
        res_resT[i]=-1
    else:
        res_resT[i]=1
# with open('分组趋势.json', 'w') as f:
#     json.dump(res_res,f)

resH={}
for i in list(set(dataH['分组信息'])):
    U[f'dataH_g{i}']=dataH[dataH['分组信息']==i]

# 遍历数据集的每一行
    for index, row in U[f'dataH_g{i}'].iterrows():
        row_sum = 0  # 用于累积每行的差值

        prev_value = None  # 用于存储前一个以"ED"开头的列的值

        # 遍历以"ED"开头的列
        for col in row.index:
            if col.startswith('HM'):
                value = row[col]
                if not pd.isna(value):  # 确保元素不是NaN
                    if prev_value is not None:
                        row_sum += (value - prev_value)  # 计算差值并累积
                    prev_value = value

        # 根据row_sum的值设置"差"列的值
        if row_sum > 0:
            U[f'dataH_g{i}'].at[index, '差'] = 1
        elif row_sum < 0:
            U[f'dataH_g{i}'].at[index, '差'] = -1
        resH[f'组{i}']=float(U[f'dataH_g{i}']['差'].sum())

res_resH = {}
for i in resH:
    if resH[i] < 0:
        res_resH[i]=-1
    else:
        res_resH[i]=1
```
