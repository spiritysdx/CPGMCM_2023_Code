```python
import numpy as np
import matplotlib.pyplot as plt

# 假设您有一个名为data的数据和string_name的标签
# 这里省略数据和标签的定义

fig, ax = plt.subplots(figsize=(15, 20))

cmap = plt.get_cmap('RdBu')

# 创建一个数组，用于设置刻度的位置，让刻度对准方格的中心
x_tick_positions = np.arange(len(string_name)) + 1.85
y_tick_positions = np.arange(len(string_name)) + 0.2

im = ax.imshow(data.values, cmap=cmap)

ax.set_xticks(x_tick_positions)
ax.set_yticks(y_tick_positions)
ax.set_xticklabels(string_name, fontsize=8, fontname='SimHei', rotation=90)
ax.set_yticklabels(string_name, fontsize=8, fontname='SimHei')

# 设置颜色条的位置和大小，并将边框设为透明
cbar = ax.figure.colorbar(im, ax=ax, shrink=0.5)
cbar.ax.set_ylabel('', rotation=-90, va="bottom")
cbar.outline.set_edgecolor('none')  # 将颜色条的边框设为透明

# 取消图的边框
ax.set_frame_on(False)

# 保存为高清图片（例如，PNG格式）
plt.savefig('q33.png', dpi=1000, bbox_inches='tight')

plt.show()
```
